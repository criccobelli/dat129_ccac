```python
# import packages that will be used 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import json

# 1001 list data scraping url
url = "http://www.rinkworks.com/checklist/list.cgi?u=crimsong&U=crimsong&p=1001movies"

page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

spans = soup.findAll('span', {'class' :['seen', 'unseen']})
lines = [span.get_text() for span in spans]
#print(spans)

# I have the span information, now I need to put it in two lists
# that will be used to search the api. These are the movie titles
# and the year each was released. I find the year by searching for
# what is in the parentheses (which is the years)
years = []
for line in lines:  
    result = ''
    paren_level = 0
    for ch in line:
        if ch == '(':
            paren_level += 1
        elif (ch == ')') and paren_level:
            paren_level -= 1
        elif paren_level:
            result += ch
    years.append(result)
    
# make new list excluding first few items which are not a year
years_list = years[10:]
print(years_list)

# next I make a new list of just the titles by excluding the years
titles = []
for i in lines:
    only_title = i[:-7]
    titles.append(only_title)
    
# as with the years list, this one had unneeded items at the beginning
title_list = titles[10:]
print(title_list)
```

```python
# my omdb api code to get information about each film
# some of the below lists are for identifying errors
falselist = []
limit_reached = []
notfound = []
search_dict = []
def movie_search(query, year):
    api_key = 'f7d58ebc'
    apiurl = 'http://www.omdbapi.com/'
    title = query.replace(' ', '+')
    final_url = apiurl + '?t=' + title + '&y=' + year + '&apikey=' + api_key
    req = requests.get(final_url)
    apiDict = json.loads(req.text)
    search_dict.append(apiDict)
    if apiDict['Response'] == 'False':
        falselist.append(query)
        if apiDict['Error'] == 'Movie not found!':
            notfound.append(query)
        if apiDict['Error'] == 'Request limit reached!':
            limit_reached.append(query)
    print(apiDict)
```

```python
# iterate through both the title and year lists to create omdb dictionaries
for x, y in zip(title_list, years_list):
    movie_search(x, y)
```

```python
# finding out more information on which films did not populate correctly
print("There are", len(falselist), "that cannot be found in total")
#print(falselist)
print("There are", len(limit_reached), "that are experiencing a request limit reached error")
print(limit_reached)
print("These films were not found for other reasons:")
print(notfound)
```
