```python
# import initial packages that will be used 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import json

# 1001 list data scraping url
url = "http://www.rinkworks.com/checklist/list.cgi?u=crimsong&U=crimsong&p=1001movies"
# open url and feed to beautifulsoup
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

# create a list which will contain all imdb links 
# from which I will retrieve the imdb ids
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
list_of_links = []
for x in links:
    if 'http://www.imdb.com/title/' in x:
        list_of_links.append(x)
print(list_of_links)

# isolating the imdb id from the link
# creating a list of only the ids
imdb_id = []
for y in list_of_links:
    onlyid = y[26:35]
    imdb_id.append(onlyid)
    
print(imdb_id)

# For some reason, one of the ids is incorrect
# and an error occurs when the API runs.
# I am finding the index of the incorrect 
# id and replacing it with the corrrect one.
index = imdb_id.index('tt3565328')
print(index)
imdb_id[49] = 'tt0020697'
```

```python
# create lists to capture ids of
# movies which cannot be found if there are any
falselist = []
limit_reached = []
notfound = []
search_dict = []
# API function which loops through the ids
# and adds them to a list called search_dict
def movie_search(id):
    api_key = 'f7d58ebc'
    apiurl = 'http://www.omdbapi.com/'
    final_url = apiurl + '?i=' + id + '&apikey=' + api_key
    req = requests.get(final_url)
    apiDict = json.loads(req.text)
    search_dict.append(apiDict)
    if apiDict['Response'] == 'False':
        falselist.append(id)
        if apiDict['Error'] == 'Movie not found!':
            notfound.append(id)
        if apiDict['Error'] == 'Request limit reached!':
            limit_reached.append(id)
    print(apiDict)
```

```python
for z in (imdb_id):
    movie_search(z)
```

```python
# testing if any of the films were not found
print(len(limit_reached))
print(notfound)
```

```python
# open csv with the movie data from the API. I named the headers
# the same as the api keys. Iterates through the search_dict list of dicts
import csv
field_names = ['Title','Year','Rated','Released','Runtime','Genre','Director','Writer','Actors','Plot','Language','Country','Awards','Poster','Ratings','Metascore','imdbRating','imdbVotes','imdbID','Type', 'totalSeasons', 'DVD','BoxOffice','Production','Website','Response']
with open('movie_data.csv','w', newline='', encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = field_names)
    writer.writeheader() 
    writer.writerows(search_dict)
```

```python
# more importing packages to be used to analysis and plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('movie_data.csv')
df.set_index(['Year', 'Title'])
df
```

```python
# some quick statistics of the list as a whole
mean_rotten = df['RottenTomatoes'].mean()
rotten_percent = mean_rotten * 100
print("The average Rotten Tomatoes score is", str(round(rotten_percent, 2)) + '%')
mean_imdb = df['imdbRating'].mean()
print("The average IMDB rating is", round(mean_imdb, 2))
mean_runtime = df['Runtime'].mean()
print("The average runtime is", round(mean_runtime, 2), "minutes")
```

```python
# In which months were the most movies released?
month_count = df['Month_Released'].value_counts()
print(month_count)
```

```python
# Which genres are most often used to describe the films?
# Since some films are described as multiple genres, 
# they must be split by commas and then counted.
print(df['Genre'].str.split(',', expand=True).stack().value_counts())
plt.rcParams["figure.figsize"] = (11,11)
df['Genre'].str.split(',', expand=True).stack().value_counts().plot(kind='pie', label='Genre')
plt.title("Most Common Genres")
```

```python
# Bar chart of all movie runtimes.
# first I set a size for the figure
plt.rcParams["figure.figsize"] = (8,8)
# then I retrieved the Runtime column
runtime = df['Runtime']
# used a bar chart to see all runtimes charted chronologically
plt.bar(df['Year'], df['Runtime'])
plt.ylabel('Minutes')
plt.xlabel('Year')
```

```python
# plotting average runtime by year
# first, group the column Year by year
df_by_year = df.groupby('Year')
# get the mean of each year for all dataframe columns
df_mean_by_year = df_by_year.mean()
# get the mean runtime of each year
df_rat_by_year = df_mean_by_year['Runtime']
#make a scatter plot of runtime by year
plt.scatter(df_rat_by_year.index, df_rat_by_year)
plt.xlabel('Year')
plt.ylabel('Runtime')
plt.title("Average Runtime By Year")
```

```python
# plotting average runtime by year
df_imdb_by_year = df_mean_by_year['imdbRating']
print(df_imdb_by_year)
plt.ylim(0, 10)
plt.scatter(df_imdb_by_year.index, df_imdb_by_year)
plt.xlabel('Year')
plt.ylabel('IMDB Rating')
```
