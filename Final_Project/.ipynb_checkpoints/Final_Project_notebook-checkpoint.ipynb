```python
# import all packages
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import csv
import requests
import json
import re
url = "https://1001films.fandom.com/wiki/The_List"

# open the webpage and send to BeautifulSoup
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

# I want to scrape the titles of each movie plus the year, these
# are contained within a 'b' tag
titles = soup.findAll('b')

# make a list of the movie titles
# I only wanted the titles so I excluded the last 6 characters
titles_list = []
for b in titles:
    title_text = b.text
    titles_list.append(title_text)
    
# the first list item was not one of the movie titles 
# so I made a new list excluding that item
total_list = titles_list[1:]
print(total_list)
```

```python
# iterate through total_list to separate the year from the title
# make new lists for each one

new_title_list = []
new_year_list = []

# just the titles    
for title in total_list:
    only_titles = title[:-7]
    new_title_list.append(only_titles)

# just the years
for year in total_list:
    year_list = year[-5:-1]
    new_year_list.append(year_list)
    
# removing parentheses if a title has them so I can search the API correctly
# some titles had foreign language titles in parentheses
final_title = []
for listing in new_title_list:  
    result = ''
    paren_level = 0
    for ch in listing:
        if ch == '(':
            paren_level += 1
        elif (ch == ')') and paren_level:
            paren_level -= 1
        elif not paren_level:
            result += ch
    final_title.append(result)
print(final_title)
```

```python
# api code for retrieving movie data
# trying to solve the problen of having some movies not search correctly in api. 
# this may be because the year on "the list" fandom website does not match omdb
# i added the ones coming back with errors to a list for if they are not
# found or if they are in the request limit

my_key = 'f7d58ebc'
falselist = []
limit_reached = []
search_dict = []
def movie_search(query, year):
    api_key = my_key
    url = 'http://www.omdbapi.com/'
    title = query.replace(' ', '+')
    final_url = url + '?t=' + title + '&y=' + year + '&apikey=' + api_key
    req = requests.get(final_url)
    apiDict = json.loads(req.text)
    if apiDict['Response'] == 'False':
        falselist.append(query)
        if apiDict['Error'] == 'Request limit reached!':
            limit_reached.append(query)
    search_dict.append(apiDict)
    print(apiDict)
```

```python
# iterate through the title and year list to create omdb dictionaries
for x, y in zip(final_title, new_year_list):
    movie_search(x, y)
```

```python
# finding out more information on which films did not populate correctly
print("There are", len(falselist), "that cannot be found in total")
print(falselist)
print("There are", len(limit_reached), "that are experiencing a request limit reached error")
print(limit_reached)
```

```python
# testing out being able to pull one type of information from the set of dicts
# finding runtime for all movies that correctly populated
for a in search_dict:
    for key, value in a.items():
        if key == 'Runtime':
            print(key+":", value)
```

```python
# create a csv file, one column for years, one for titles
fields = ['Year', 'Title']
with open('moviedata.csv', 'w', newline='', encoding="utf-8") as f:
    wr = csv.writer(f)
    wr.writerow(["Year", "Title"])
    wr.writerows(zip(new_year_list, final_title))
f.close()
```
