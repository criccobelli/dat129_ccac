```python
# import packages that will be used 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import json

# 1001 list data scraping url
url = "http://www.rinkworks.com/checklist/list.cgi?u=crimsong&U=crimsong&p=1001movies"

page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

#spans = soup.findAll('a')
#lines = [span.get_text() for span in spans]
#print(lines)
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
list_of_links = []
for x in links:
    if 'http://www.imdb.com/title/' in x:
        list_of_links.append(x)
#print(list_of_links)

imdb_id = []
for y in list_of_links:
    onlyid = y[26:35]
    imdb_id.append(onlyid)
#print(imdb_id)

index = imdb_id.index('tt3565328')
print(index)
# for some reason, one of the ids is incorrect
# I can manually replace it with the correct one
imdb_id[49] = 'tt0020697'
```

```python
falselist = []
limit_reached = []
notfound = []
search_dict = []
def movie_search(id):
    api_key = 'f7d58ebc'
    apiurl = 'http://www.omdbapi.com/'
    final_url = apiurl + '?i=' + id + '&apikey=' + api_key
    req = requests.get(final_url)
    apiDict = json.loads(req.text)
    search_dict.append(apiDict)
    if apiDict['Response'] == 'False':
        falselist.append(id)
        if apiDict['Error'] == 'Movie not found!':
            notfound.append(id)
        if apiDict['Error'] == 'Request limit reached!':
            limit_reached.append(id)
    print(apiDict)
```

```python
for z in (imdb_id):
    movie_search(z)
```

```python
# make a new list of dicts that don't include the request limit ones
# in order to practice getting the information into a csv
good_search = search_dict[0:195]
print(good_search)
```

```python
# open csv with the data I have so far. I'll use the same
# headers that the api uses as keys
import csv
field_names = ['Title','Year','Rated','Released','Runtime','Genre','Director','Writer','Actors','Plot','Language','Country','Awards','Poster','Ratings','Metascore','imdbRating','imdbVotes','imdbID','Type','DVD','BoxOffice','Production','Website','Response']
with open('partial_data.csv','w', newline='', encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = field_names)
    writer.writeheader() 
    writer.writerows(good_search)
```
