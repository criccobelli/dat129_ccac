```python
# import all packages
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import csv
import requests
import json
url = "https://1001films.fandom.com/wiki/The_List"

# open the webpage and send to BeautifulSoup
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

# I want to scrape the titles of each movie plus the year, these
# are contained within a 'b' tag
titles = soup.findAll('b')

# make a list of the movie titles
# I only wanted the titles so I excluded the last 6 characters
titles_list = []
for b in titles:
    title_text = b.text
    titles_list.append(title_text)
    
# the first list item was not one of the movie titles 
# so I made a new list excluding that item
total_list = titles_list[1:]
print(total_list)
```

```python
# iterate through total_list to separate the year from the title
# make new lists for each one

new_title_list = []
new_year_list = []

# just the titles    
for title in total_list:
    only_titles = title[:-7]
    new_title_list.append(only_titles)

# just the years
for year in total_list:
    year_list = year[-5:-1]
    new_year_list.append(year_list)
    
print(new_title_list)
print(new_year_list)
    
```

```python
# create a csv file, one column for years, one for titles
fields = ['Year', 'Title']
with open('movie_data.csv', 'w', newline='', encoding="utf-8") as f:
    wr = csv.writer(f)
    wr.writerow(["Year", "Title"])
    wr.writerows(zip(new_year_list, new_title_list))

# create pandas dataframe
#df = pd.DataFrame(new_title_list, new_year_list)
#print(df)
```
